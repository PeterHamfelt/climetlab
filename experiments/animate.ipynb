```python
import climetlab as cml
source = cml.load_source(
    "cds",
    "reanalysis-era5-single-levels",
    variable="2t",
    product_type="reanalysis",
    area=[50, -50, 20, 50],
    date="2012-12-12",
    time='all',
)
```

```python
p = cml.new_plot(animate=True, fps=5)
for f in source:
    p.plot_map(f)

p.show()
```
